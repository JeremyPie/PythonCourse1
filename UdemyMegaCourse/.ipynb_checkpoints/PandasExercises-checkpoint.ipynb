```python
df2 = pd.read_csv("supermarkets-commas.csv")
df2
```

```python
df2['Address'] = df2['Address'] + ', ' + df2['City'] + ', ' + df2['State'] + ', ' + df2['Country']
df2
```

```python
for adress in df2.loc[:, 'Address']:
    print(nom.geocode(adress).latitude)
```

```python
df2["Coordinates"] = df2["Address"].apply(nom.geocode)
df2
```

```python
df2["Latitude"] = df2["Coordinates"].apply(lambda x: x.latitude)
df2
```

```python
df2["Longitude"] = df2["Coordinates"].apply(lambda x: x.longitude)
df2
```
